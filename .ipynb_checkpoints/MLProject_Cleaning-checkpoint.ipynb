```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv', index_col=0)
```

```python
train1 = train.copy()
```

```python
##just a test 
train1['LotFrontage'].replace(np.nan,'No',inplace=True)
```

```python
#ob with a basement, but shows "no basement" in the BsmtExposure col NEED TO IMPUTE
train1[train1['BsmtExposure'].isnull() & train1['BsmtQual'].notnull()]
```

```python
#ob with a BsmtFinSF2 but shows "no basement" in the BsmtFinType2 col NEED TO IMPUTE
train1[train1['BsmtFinType2'].isnull() & train1['BsmtFinSF2']!=0]
```

```python
#cols for which NA will be replaced with "No" because NA is a true category thus need to change 
#to "No" so Pandas does not pick it up as np.NaN
FixNA=['Alley', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'FireplaceQu', 'GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC', 'Fence', 'MiscFeature']
```

```python
#list of categorical variables 
cat_variables=['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Foundation', 'BsmtExposure', 'BsmtFinType2', 'Heating', 'CentralAir', 'Electrical', 'GarageType', 'MiscFeature', 'SaleType', 'SaleCondition']
```

```python
#dummify categorical nominal and binary variables
for v in cat_variables:
    temp = pd.get_dummies(train1[v],prefix=v, prefix_sep='_')
    temp = temp.drop(temp.iloc[:,0:1],axis=1)
    train1 = pd.concat([train1.drop(v,axis=1),temp],axis=1)
```
